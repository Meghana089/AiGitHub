```python
# Generate basic data types and variables in Python

# Numeric types
integer_var = 42
float_var = 3.14
complex_var = 2 + 3j

# String type
string_var = "Hello, Python!"

# Boolean type
bool_var = True

# List (mutable, ordered collection)
list_var = [1, 2, 3, 4, 5]

# Tuple (immutable, ordered collection)
tuple_var = (10, 20, 30)

# Dictionary (key-value pairs)
dict_var = {"name": "Alice", "age": 25, "city": "New York"}

# Set (unordered, unique elements)
set_var = {1, 2, 3, 4, 5}

# None type
none_var = None

# Display all variables
print(f"Integer: {integer_var}, Type: {type(integer_var)}")
print(f"Float: {float_var}, Type: {type(float_var)}")
print(f"Complex: {complex_var}, Type: {type(complex_var)}")
print(f"String: {string_var}, Type: {type(string_var)}")
print(f"Boolean: {bool_var}, Type: {type(bool_var)}")
print(f"List: {list_var}, Type: {type(list_var)}")
print(f"Tuple: {tuple_var}, Type: {type(tuple_var)}")
print(f"Dictionary: {dict_var}, Type: {type(dict_var)}")
print(f"Set: {set_var}, Type: {type(set_var)}")
print(f"None: {none_var}, Type: {type(none_var)}")
```

```python
# Control flow with if statements

# Example 1: Check integer value
if integer_var > 40:
    print(f"{integer_var} is greater than 40")
else:
    print(f"{integer_var} is not greater than 40")

# Example 2: Check string length
if len(string_var) > 10:
    print(f"String '{string_var}' is long")
else:
    print(f"String '{string_var}' is short")

# Example 3: Check if list is not empty
if list_var:
    print(f"List has {len(list_var)} elements")

# Example 4: Check boolean value
if bool_var:
    print("Boolean is True")

# Example 5: Check dictionary key existence
if "age" in dict_var:
    print(f"Age in dictionary: {dict_var['age']}")

# Example 6: Multiple conditions (if-elif-else)
if float_var > 5:
    print("Float is greater than 5")
elif float_var > 2:
    print("Float is between 2 and 5")
else:
    print("Float is 2 or less")
```

```python
# Control flow with else and elif statements

# Example 1: Check if integer is even or odd
if integer_var % 2 == 0:
    print(f"{integer_var} is even")
else:
    print(f"{integer_var} is odd")

# Example 2: Categorize float value
if float_var < 1:
    print("Float is less than 1")
elif float_var < 3:
    print("Float is between 1 and 3")
else:
    print("Float is 3 or greater")

# Example 3: Check set operations
if set_var == {1, 2, 3, 4, 5}:
    print("Set matches expected values")
else:
    print("Set does not match")

# Example 4: Check tuple length
if len(tuple_var) > 5:
    print("Tuple is long")
else:
    print("Tuple is short")

# Example 5: Complex nested conditions
if dict_var.get("age", 0) >= 25:
    print(f"{dict_var['name']} is 25 or older")
else:
    print(f"{dict_var['name']} is younger than 25")
```

```python
# Control flow with elif statements

# Example 1: Categorize integer into ranges
if integer_var < 20:
    print(f"{integer_var} is less than 20")
elif integer_var < 40:
    print(f"{integer_var} is between 20 and 40")
elif integer_var < 60:
    print(f"{integer_var} is between 40 and 60")
else:
    print(f"{integer_var} is 60 or greater")

# Example 2: Check list length with elif
if len(list_var) < 3:
    print("List is small")
elif len(list_var) < 7:
    print("List is medium")
else:
    print("List is large")

# Example 3: Categorize string length
if len(string_var) < 5:
    print("String is very short")
elif len(string_var) < 10:
    print("String is short")
elif len(string_var) < 15:
    print("String is medium")
else:
    print("String is long")

# Example 4: Check dictionary age with elif
age = dict_var.get("age", 0)
if age < 18:
    print("Minor")
elif age < 65:
    print("Adult")
else:
    print("Senior")
```

```python
# Example 1: Iterate through list
for item in list_var:
    print(f"List item: {item}")

# Example 2: Iterate through tuple
for num in tuple_var:
    print(f"Tuple item: {num}")

# Example 3: Iterate through set
for element in set_var:
    print(f"Set element: {element}")

# Example 4: Iterate through dictionary keys
for key in dict_var:
    print(f"Dictionary key: {key}, value: {dict_var[key]}")

# Example 5: Iterate with index using enumerate
for index, item in enumerate(list_var):
    print(f"Index {index}: {item}")

# Example 6: Iterate through string characters
for char in string_var:
    print(char, end=" ")
print()  # New line

# Example 7: For loop with range
for i in range(5):
    print(f"Number: {i}")

# Example 8: For loop with conditional
for num in list_var:
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
# While loop examples

# Example 1: Simple while loop with counter
counter = 0
while counter < 5:
    print(f"Counter: {counter}")
    counter += 1

# Example 2: While loop with list
index = 0
while index < len(list_var):
    print(f"List item: {list_var[index]}")
    index += 1

# Example 3: While loop with condition on variable
remaining = integer_var
while remaining > 0:
    print(f"Remaining: {remaining}")
    remaining -= 10

# Example 4: While loop with break
count = 0
while True:
    if count == 3:
        break
    print(f"Count: {count}")
    count += 1

# Example 5: While loop with continue
num = 0
while num < 5:
    num += 1
    if num == 3:
        continue
    print(f"Number: {num}")
```

```python
# Simulate do-while loop using while True and break

# Example 1: Simple do-while simulation
user_input = None
while True:
    print("This executes at least once")
    user_input = input("Enter 'quit' to exit: ")
    if user_input == 'quit':
        break

# Example 2: do-while with counter
attempt = 0
while True:
    print(f"Attempt: {attempt}")
    attempt += 1
    if attempt >= 3:
        break

# Example 3: do-while with list iteration
index = 0
while True:
    print(f"List item: {list_var[index]}")
    index += 1
    if index >= len(list_var):
        break
```
